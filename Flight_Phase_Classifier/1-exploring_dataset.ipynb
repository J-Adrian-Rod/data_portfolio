```python
file_path = 'raw_data/Tail_681_1/681200103250318.mat'
mat = scipy.io.loadmat(file_path)

type_counts = {}
print(" Variables inside MAT file:")

for k, v in mat.items():
    v_type = type(v).__name__
    v_shape = getattr(v, 'shape', None)
    print(f" {k:<20} | type: {v_type:<15} | shape: {v_shape}")
    type_counts[v_type] = type_counts.get(v_type, 0) + 1

print(" Summary of types:")
for t, count in type_counts.items():
    print(f"   {t}: {count} variables")
```
